```python
########################################
#Faktoriyel
########################################
def faktoriyel(sayi):
    fact = 1 
    if sayi < 0: 
        print ("Sayi sifirdan kücük olamaz")
    else:
        for i in range(1,sayi + 1): 
            fact = fact*i 
    return fact
########################################
#Permutasyon
########################################
def permutasyon(sboncuk,bboncuk):
    perm = 0
    perm = faktoriyel(sboncuk+bboncuk)/(faktoriyel(sboncuk)*faktoriyel(bboncuk))
    return perm
########################################
#Kombinasyon
########################################
def kombinasyon(toplam_sayi,ssayi):
    komb = 0
    if (ssayi>toplam_sayi):
        print ("Lütfen eleman sayısından küçük bir seçim sayısı giriniz:")
    else:
        komb = faktoriyel(toplam_sayi)/(faktoriyel(toplam_sayi-ssayi)*faktoriyel(ssayi))
    return komb

p = 0.90
n = 5
q = 0.10
def kesikli_olasilik(istenilen_durum):

    p_temp = p**istenilen_durum
    q_temp = q**(n-istenilen_durum)
    result =kombinasyon(n,istenilen_durum)*p_temp*q_temp
    return result
```

```python
print(kesikli_olasilik(5))
```

```python
print(kesikli_olasilik(2))
```

```python
print(kesikli_olasilik(4)+kesikli_olasilik(5))
```

```python
print(kesikli_olasilik(0)+kesikli_olasilik(1)+kesikli_olasilik(2))
```
